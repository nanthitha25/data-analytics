```python
import pandas as pd
```

```python
data = pd.read_csv('/content/unclean.csv')
data
```

```python
df = pd.DataFrame(data)
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df['Age'] = df['Age'].replace({'nineteen': 19, 'Nineteen': 19})
df
```

```python
df['Age'] = df['Age'].fillna('18')
print (df)
```

```python
df['City'].fillna("Unknown", inplace=True)
print(df)
```

```python
df['Order_Date'].fillna(pd.Timestamp('2024-01-01'), inplace=True)
print(df)
```

```python
df['Product'].fillna("cupcake", inplace=True)
df
```

```python
df['Price'].fillna("unknown", inplace=True)
df
```

```python
df['Quantity']= df['Quantity'].replace({'one' : 1})
df['Quantity'].fillna("1", inplace=True)
df
```

```python
df['Rating'].fillna('0.1' ,inplace = True)
df
```

```python
df
```

```python
df['Gender'].fillna('prefer not to say' , inplace = True)
df
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
import matplotlib.pyplot as plt

# Data for Figure 7
ai_systems = ['API Systems', 'LLM Systems', 'Autonomous Agents']
accuracy = [90, 92, 90]

# Create bar chart
plt.figure(figsize=(8, 5))
bars = plt.bar(ai_systems, accuracy)

# Add value labels on bars
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height + 0.5,
        f'{height}%',
        ha='center',
        va='bottom'
    )

# Labels and title
plt.ylabel('Detection Accuracy (%)')
plt.xlabel('AI System Type')
plt.title('Figure 7. Vulnerability Detection Accuracy Across AI System Types')
plt.ylim(0, 100)

# Grid for readability
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Save figure for journal
plt.tight_layout()
plt.savefig('figure7_accuracy_chart.png', dpi=300)
plt.show()
```
