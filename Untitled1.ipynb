```python
from matplotlib.lines import lineStyles
import matplotlib.pyplot as plt

x=[1,2,4,67,87]
y=[2,4,5,6,78]

plt.plot(x,y,color='red',linestyle='--',marker='o',markerfacecolor='b',markeredgecolor='y')
plt.title("Line chart")
plt.xlabel="x"
plt.ylabel="y"
plt.show()
```

```python
from matplotlib.lines import lineStyles
import matplotlib.pyplot as plt

a=[1,2,4,67,87]
b=[2,4,5,6,78]
plt.plot(x,y,label = "salary")
plt.plot(a,b,label ='growth')
plt.title ("multi-Line chart")
plt.legend()
plt.show()
```
